```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/alletsc/wine-tech-challenge-fiap/main/assets/data/exported_data2007_2022.csv"

exported_data = pd.read_csv(url)

exported_data.head()
```

```python
# Tipando dados
exported_data.dtypes
```

```python
exported_data["Quantidade (kg/L)"] = exported_data["Quantidade (kg/L)"].astype(float)
exported_data["Valor em US$"] = exported_data["Valor em US$"].astype(float)

exported_data.dtypes
```

```python
exported_data.head()
```

```python
# Grafico 1: exportação de vinho por quantidade (kg/L) por ano
exported_data.groupby("Ano")["Quantidade (kg/L)"].sum().plot(kind="bar", title="Exportação de vinho por quantidade (kg/L) por ano")
```

```python
# Grafico 2: exportação de vinho por país de destino 15 primeiros
exported_data.groupby("País de destino")["Quantidade (kg/L)"].sum().nlargest(15).plot(kind="bar", title="Exportação de vinho por país de destino 15 primeiros")
```

```python
# Grafico 2: exportação de vinho por país de destino 15 primeiros por valor
exported_data.groupby("País de destino")["Valor em US$"].sum().nlargest(15).plot(kind="bar", title="Exportação de vinho por país de destino 15 primeiros por valor")
```

```python
# Grafico 3: exportação de vinho por país de destino 15 últimos
exported_data.groupby("País de destino")["Quantidade (kg/L)"].sum().nsmallest(15).plot(kind="bar", title="Exportação de vinho por país de destino 15 últimos")
```
